# transform

```python
customers_df = orders_df[["CustomerName", "State", "City"]].drop_duplicates().reset_index(drop=True)
customers_df["Customer ID"] = customers_df.index + 1

customers_df = customers_df[["Customer ID", "CustomerName", "State", "City"]]
customers_df
```

```python
old_to_new = {j: i+1 for i, j in enumerate(orders_df['Order ID'])}
old_to_new
```

```python
orders_df["Customer ID"] = orders_df.merge(customers_df, on=["CustomerName", "State", "City"])["Customer ID"]
orders_table = orders_df[["Order ID", "Order Date", "Customer ID"]]
orders_table["Order ID"] = orders_table.index + 1

orders_table
```

```python
categories_df = order_details_df[["Category"]].drop_duplicates().reset_index(drop=True)
categories_df["Category ID"] = categories_df.index + 1

categories_df = categories_df[["Category ID", "Category"]]
categories_df
```

```python
subcategories_df = order_details_df[["Sub-Category", "Category"]].drop_duplicates().reset_index(drop=True)
subcategories_df = subcategories_df.merge(categories_df, on="Category")
subcategories_df["Sub-Category ID"] = subcategories_df.index + 1
subcategories_df = subcategories_df.drop('Category', axis='columns')

subcategories_df = subcategories_df[["Sub-Category ID", "Sub-Category", "Category ID"]]
subcategories_df
```

```python
order_details_table = order_details_df.merge(
    subcategories_df[["Sub-Category", "Sub-Category ID"]],
    on="Sub-Category"
)
order_details_table["Order ID"] = order_details_table["Order ID"].map(old_to_new)
order_details_table["Detail ID"] = order_details_table.index + 1

order_details_table = order_details_table[["Detail ID", "Order ID", "Amount", "Profit", "Quantity", "Sub-Category ID"]]
order_details_table
```

```python
sales_targets_table = sales_targets_df.merge(
    categories_df, left_on="Category", right_on="Category"
)
sales_targets_table = sales_targets_table[["Month of Order Date", "Category ID", "Target"]]
sales_targets_table["Target ID"] = sales_targets_table.index + 1

sales_targets_table = sales_targets_table[["Target ID", "Month of Order Date", "Category ID", "Target"]]
sales_targets_table
```
