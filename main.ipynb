```python
import pandas as pd

pd.options.display.float_format = "{:,.2f}".format
```

# read data

```python
orders_df = pd.read_csv('./datasorce/List of Orders.csv')
order_details_df = pd.read_csv('./datasorce/Order Details.csv')
sales_targets_df = pd.read_csv('./datasorce/Sales target.csv')
```

```python
orders_df.info()
```

```python
orders_df = orders_df.dropna();
orders_df
```

```python
CustomerNameUnique = orders_df['CustomerName'].unique()
print(CustomerNameUnique, len(CustomerNameUnique))
```

```python
MapIDCustomerName = [ {'id': f"C-{i+101:03}", 'name': j} for i, j in enumerate(CustomerNameUnique) ]
MapIDCustomerName
```

```python
StateUnique = orders_df['State'].unique()
print(StateUnique, len(StateUnique))
```

```python
CityUnique = orders_df['City'].unique()
print(CityUnique, len(CityUnique))
```

```python
order_details_df.info()
```

```python
CategoryUnique = order_details_df['Category'].unique()
print(CategoryUnique, len(CategoryUnique))
```

```python
SubCategoryUnique = order_details_df['Sub-Category'].unique()
print(SubCategoryUnique, len(SubCategoryUnique))
```

```python
sales_targets_df.info()
```

# transform

```python
name_to_id = {customer['name']: customer['id'] for customer in MapIDCustomerName}
name_to_id
```

```python
customers_df = orders_df[["CustomerName", "State", "City"]].drop_duplicates().reset_index(drop=True)

customers_df["Customer ID"] = customers_df['CustomerName'].map(name_to_id)

customers_df = customers_df[["Customer ID", "CustomerName", "State", "City"]]
customers_df
```

```python
orders_df["Customer ID"] = customers_df["Customer ID"]
orders_table = orders_df[["Order ID", "Order Date", "CustomerName"]]

orders_table["Customer ID"] = orders_table['CustomerName'].map(name_to_id)

orders_table = orders_table.drop('CustomerName', axis='columns')

orders_table
```

```python
categories_df = order_details_df[["Category"]].drop_duplicates().reset_index(drop=True)
categories_df["Category ID"] = [f"CAT-{i:02}" for i in range(1, len(categories_df) + 1)]

categories_df = categories_df[["Category ID", "Category"]]
categories_df
```

```python
subcategories_df = order_details_df[["Sub-Category", "Category"]].drop_duplicates().reset_index(drop=True)
subcategories_df = subcategories_df.merge(categories_df, on="Category")
subcategories_df["Sub-Category ID"] = [f"SC-{i:02}" for i in range(1, len(subcategories_df) + 1)]
subcategories_df = subcategories_df.drop('Category', axis='columns')

subcategories_df = subcategories_df[["Sub-Category ID", "Sub-Category", "Category ID"]]
subcategories_df
```

```python
order_details_table = order_details_df.merge(
    subcategories_df[["Sub-Category", "Sub-Category ID"]],
    on="Sub-Category"
)
order_details_table["Detail ID"] = [f"D-{i:04}" for i in range(1, len(order_details_table) + 1)]
order_details_table = order_details_table[["Detail ID", "Order ID", "Amount", "Profit", "Quantity", "Sub-Category ID"]]

order_details_table
```

```python
sales_targets_table = sales_targets_df.merge(
    categories_df, left_on="Category", right_on="Category"
)
sales_targets_table = sales_targets_table[["Month of Order Date", "Category ID", "Target"]]
sales_targets_table["Target ID"] = [f"T-{i:04}" for i in range(1, len(sales_targets_table) + 1)]

sales_targets_table = sales_targets_table[["Target ID", "Month of Order Date", "Category ID", "Target"]]
sales_targets_table
```
