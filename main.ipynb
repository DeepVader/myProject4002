# transform

```python
name_to_id = {customer['name']: customer['id'] for customer in MapIDCustomerName}
name_to_id
```

```python
customers_df = orders_df[["CustomerName", "State", "City"]].drop_duplicates().reset_index(drop=True)

customers_df["Customer ID"] = customers_df['CustomerName'].map(name_to_id)

customers_df
```

```python
orders_df["Customer ID"] = customers_df["Customer ID"]
orders_table = orders_df[["Order ID", "Order Date", "CustomerName"]]

orders_table["Customer ID"] = orders_table['CustomerName'].map(name_to_id)

orders_table = orders_table.drop('CustomerName', axis='columns')
orders_table
```

```python
categories_df = order_details_df[["Category"]].drop_duplicates().reset_index(drop=True)
categories_df["Category ID"] = range(1, len(categories_df) + 1)
categories_df
```

```python
subcategories_df = order_details_df[["Sub-Category", "Category"]].drop_duplicates().reset_index(drop=True)
subcategories_df = subcategories_df.merge(categories_df, on="Category")
subcategories_df["Sub-Category ID"] = [f"SC-{i:04}" for i in range(1, len(subcategories_df) + 1)]
subcategories_df
```
